# Step 1: Preprocess the Data

```python
import pandas as pd

# Load the CIC-IDS-2017 dataset
# Replace 'your_dataset_file.csv' with the actual file name of your CIC-IDS-2017 dataset
cic_ids_2017_path = '/content/drive/My Drive/CIC - IDS 2017/Monday-WorkingHours.pcap_ISCX.csv'  # Example file
cic_ids_2017_data = pd.read_csv(cic_ids_2017_path)

# Display the first few rows of the dataset
print("CIC-IDS-2017 Dataset:")
print(cic_ids_2017_data.head())
```

Step 1: Check for Infinite Values

```python
import numpy as np

# Replace infinite values with NaN
cic_ids_2017_data.replace([np.inf, -np.inf], np.nan, inplace=True)

# Optionally, fill NaN values with zero or the mean of the column
cic_ids_2017_data.fillna(0, inplace=True)
# Alternatively, you can fill NaN with the mean of each column
# cic_ids_2017_data.fillna(cic_ids_2017_data.mean(), inplace=True)
```

Step 2: Scale the Features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features = cic_ids_2017_data.drop(' Label', axis=1)
features_scaled = scaler.fit_transform(features)
```

Step 3: Continue with Data Splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features_scaled, cic_ids_2017_data[' Label'], test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Handle missing values if any
cic_ids_2017_data.fillna(0, inplace=True)

# Encode categorical variables
label_encoder = LabelEncoder()
cic_ids_2017_data[' Label'] = label_encoder.fit_transform(cic_ids_2017_data[' Label'])

# Normalize the data
scaler = StandardScaler()
features = cic_ids_2017_data.drop(' Label', axis=1)
features_scaled = scaler.fit_transform(features)

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features_scaled, cic_ids_2017_data[' Label'], test_size=0.2, random_state=42)
```

# Step 2: Build and Train the Model

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# Define the Autoencoder model
input_dim = X_train.shape[1]
encoding_dim = 14  # You can adjust this value

input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation="relu")(input_layer)
decoder = Dense(input_dim, activation="sigmoid")(encoder)

autoencoder = Model(inputs=input_layer, outputs=decoder)

# Compile the model
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
autoencoder.fit(X_train, X_train, epochs=50, batch_size=256, shuffle=True, validation_split=0.2)
```

```python
# Predict on test data
predictions = autoencoder.predict(X_test)

# Calculate reconstruction error for each sample
mse = tf.keras.losses.MeanSquaredError(reduction=tf.keras.losses.Reduction.NONE)
reconstruction_error = mse(X_test, predictions).numpy()

# Determine a threshold for anomaly detection
threshold = 0.5  # This is an example; you may need to adjust based on your data

# Classify each sample as anomaly or normal
anomalies = reconstruction_error > threshold

# Evaluate the model
from sklearn.metrics import classification_report

print(classification_report(y_test, anomalies))
```
